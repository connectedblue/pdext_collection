```python
from random import randint
from datetime import datetime, timedelta
import pandex as pd
import numpy as np


# pd.ext.import_extension('. -> year_heatmap')
pd.ext.reinstall_extension('year_heatmap')
```

```python
date_today = datetime.now()
days = pd.date_range(date_today+timedelta(randint(10,10)), 
                     date_today + timedelta(randint(600,800)), freq='D')

np.random.seed(seed=1111)
df = pd.DataFrame({'date': days, 
                   'Sales Volume': np.random.randint(300, high=1800, size=len(days)),
                   'Discounts Applied': np.random.randint(300, high=1800, size=len(days)),
                  })
df = df.set_index('date')
```

```python
fig = df.ext.year_heatmap()
```
